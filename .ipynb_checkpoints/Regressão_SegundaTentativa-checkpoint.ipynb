## Normalização:

```python
# Criando um df pra ultima normalização:
df_normalizados = df_treino[var_relevantes].copy() # To criando esse outro dataframe pra ficar libre pra testar normalizações

# Normalizando as Multicategoricas:
for col in tqdm(df_treino[var_relevantes].columns):
    if col != 'Crack/Similares':
        df_transicao = pd.get_dummies(df_normalizados[col])
        df_normalizados = pd.concat([df_normalizados, df_transicao], axis=1)
        df_normalizados = df_normalizados.drop([col], axis=1)
df_normalizados.head(5)
```

```python
# Normalização do Crack
df_normalizados['Crack/Similares'] = df_normalizados['Crack/Similares'].replace('Não', 0)
df_normalizados['Crack/Similares'] = df_normalizados['Crack/Similares'].replace('Sim', 1)
```

```python
df_normalizados
```

# Regressão:
Vou usar dois modelos para a Regressão, regressão linear e a arvore de Decisões e comparar qual que é a melhor

```python
# Criar lista de valores
valores = ['Dados inconclusios']

# Dividir o dataframe em dois
df_teste = df_normalizados[df_normalizados['Crack/Similares'].isin(valores)]
df_treino = df_normalizados[~df_normalizados['Crack/Similares'].isin(valores)]
```

```python
#Definindo os eixos x e y:
x = df_treino.drop(['Crack/Similares'], axis=1) # Porque como ela será o output, essa variável não pode estar no conjunto x
y = df_treino['Crack/Similares'] # Essa variável precisa estar destacada aqui, no eixo y.


#Dividindo o dataset entre treino e teste:
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
#Regressão Linear:
lin_reg = LinearRegression()
lin_reg.fit(x_train,y_train)

#Árvore de Decisões:
rf_reg = RandomForestRegressor()
rf_reg.fit(x_train,y_train)
```

```python
#x_test:
test_pred_lin = lin_reg.predict(x_test)
test_pred_rf = rf_reg.predict(x_test)

#y_test - Regressão Linear:
r2_lin = metrics.r2_score(y_test, test_pred_lin)
rsme_lin = np.sqrt(metrics.mean_squared_error(y_test, test_pred_lin))
print('Regressão Linear:')
print(f'R² da Regressão Linear:{r2_lin}')
print(f'RSME da Regressão Linear:{rsme_lin}')
print('\n')

#y_test - Random Forest:
r2_rf = metrics.r2_score(y_test, test_pred_rf)
rsme_rf = np.sqrt(metrics.mean_squared_error(y_test, test_pred_rf))
print('Arvore de Decisões:')
print(f'R² do Random Forest:{r2_rf}')
print(f'RSME do Random Forest:{rsme_rf}')
print('\n')
```

```python
#Preparando o DataFrame para o Gráfico:
df_gráfico = pd.DataFrame()
df_gráfico['y_teste']= y_test
df_gráfico['y_previsão_rf']= test_pred_rf
df_gráfico['y_previsão_lin']= test_pred_lin
df_gráfico = df_gráfico.reset_index(drop=True)
df_gráfico
```

## Visualizações do Modelo

```python
#Gráfico de Comparação:
fig = plt.figure(figsize=(100,10))
sns.lineplot(data=df_gráfico)
plt.legend(['Teste', 'Previsão da Regressão Linear','Previsão da Árvore de Decisões'])
plt.title('Gráfico de comparação entre o teste e as previsões')
plt.xlabel('Index - Valores no tempo')
plt.ylabel('Valor de Saída')
plt.grid(True)
print('Clique na imagem para melhor visualização')
plt.show()
```

### Variáveis relevantes:

```python
# Relevancia pela Arvore de Decisões:
features_relevantes = pd.DataFrame(rf_reg.feature_importances_,x_train.columns)
plt.figure(figsize=(80,12))
plt.grid(True)
plt.xticks(rotation=75)
sns.barplot(x=features_relevantes.index, y=features_relevantes[0])
plt.show()
```

```python
# Checando a importância relativa de cada variável para a previsão:
features_relevantes_lr = pd.DataFrame(lin_reg.coef_,x_train.columns)
plt.figure(figsize=(80,12))
plt.grid(True)
plt.xticks(rotation=83)
sns.barplot(x=features_relevantes_lr.index, y=features_relevantes_lr[0])
plt.title("Relevância das Variáveis no Modelo de Regressão Linear")
plt.savefig('VarRelevance_LinReg.png', bbox_inches='tight')
plt.show()
```

### Resíduos:

```python
# Residos da árvore de decisões:
y_true = df_treino['Crack/Similares']
y_pred = df_gráfico['y_previsão_rf']
residues_RF = y_true - y_pred
print(residues_RF.sum())

# Plotar os resíduos
plt.figure(figsize=(20,10))
plt.grid(True)
sns.histplot(residues_RF, bins=50, color='blue')
plt.title("Resíduos da Árvore de Decisão")
plt.xlabel("Valor dos Resíduos")
plt.ylabel("Frequência")
plt.savefig('Resíduos_RF.png', bbox_inches='tight')
plt.show()
```

```python
# Resíduos da Regressão linear:
y_true = df_treino['Crack/Similares']
y_pred = df_gráfico['y_previsão_lin']
residues_linReg = y_true - y_pred
print(residues_linReg.sum())

# Plotar os resíduos
plt.figure(figsize=(20,10))
plt.grid(True)
sns.histplot(residues_linReg, bins=50, color='red')
plt.title("Resíduos da Regressão Linear")
plt.xlabel("Valor dos Resíduos")
plt.ylabel("Frequência")
plt.savefig('Resíduos_LinReg.png', bbox_inches='tight')
plt.show()
```

```python
# Calcular os resíduos
y_true = residues_RF
y_pred = residues_linReg
residues_final = y_true - y_pred

# Plotar os resíduos
plt.figure(figsize=(20,10))
plt.grid(True)
sns.histplot(residues_final, bins=50, color='purple')
plt.title("Comparação dos resíduos da Árvore de Decisões e Regressão Linear")
plt.xlabel("Valor dos Resíduos")
plt.ylabel("Frequência")
plt.savefig('Resíduos_LinReg_RF.png', bbox_inches='tight')
plt.show()
```
